```python
import numpy as np
import cv2
from IPython.display import display
from PIL import Image
```

```python
image = cv2.imread('/Users/dilshod/Downloads/cat.jpg')
cv2.circle(image,(300,680), 70, (0,0,255), 5)
cv2.circle(image,(690,620), 60, (0,0,255), 5)
pts = np.array([[100,380],[70,200],[110,60],[200,30],[350,50],[450,270]], np.int32)
pts = pts.reshape((-1,1,2))
cv2.polylines(image,[pts],True,(0,255,0))
display(Image.fromarray(image))
cv2.imwrite('cat_processed.jpg', image)
```
